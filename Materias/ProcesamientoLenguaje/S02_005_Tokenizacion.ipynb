Comencemos con un texto inicial para trabajar

```
texto = 'El jefe del fabricante de vehículos eléctricos Tesla, Elon Musk, dijo que va a retomar la producción en la planta de la compañía en California, "Tesla va a reiniciar la producción hoy en contra de las reglas del condado de Alameda", dijo Musk en un tuit.'
```

Para segmentar en "palabras" recordemos que python ofrece una herramienta sumamente sencilla.

```
palabras = texto.split()

print(palabras)
```

Todo se ve bien a primera vista, pero no todo es lo que aparenta. Intentemos mostrar el vocabulario del texto.

```
vocabulario = set(palabras)

sorted(vocabulario)
```

Se pueden ver cosas raras, aparecen repetidas palabras como "Tesla" y "Musk". No hay un error en el conjunto, son "palabras" diferentes por culpa de los signos de puntuación. La función `s.split()` divide en espacios pero no toma ninguna consideración especial para la puntuación, que no forma parte de las palabras tampoco.

Una solución simplista sería quitar la la puntuación antes de hacer la segmentación.

```
separadores = ["(",")",",",".",";",":","\"","¿","?","¡","!"]

for separador in separadores:
  texto = texto.replace(separador," ")
```

```
palabras = texto.split()
vocabulario = set(palabras)

sorted(vocabulario)
```

Eso ya esta mucho mejor, pero con eso no se terminan los problemas.

Veamos ahora cómo utilizar herramientas diseñadas específicamente para la tokenización.

```
import nltk
```

NLTK es una librería con muchas herramientas para el procesamiento de lenguaje natural, pero no tiene todo pre-cargado. Para poder hacer uso de tokenizadores, se necesita descargarlos.

```
nltk.download()
```

El paquete que se necesita descargar es el paquete `punkt`. Asi que al módulo de descarga se le deba dar la instrucción `d` y después la palabra `punkt` cuando pregunte por el paquete.

Si hacen esto de forma local, el menú será distinto, aparecerá una ventana en donde podrán buscar el paquete con interfaz gráfica.
![](https://raw.githubusercontent.com/apimentelaUP/corpora/master/punkt.png)

```
tokens = nltk.word_tokenize(texto,"spanish")

print(tokens)
```
