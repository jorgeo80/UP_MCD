Vamos a cargar un texto para los ejercicios de stemming. Podemos encontrar el texto en la liga: 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/M%C3%A9xico_texto.txt

```
import requests

url = 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/M%C3%A9xico_texto.txt'
page = requests.get(url)
texto = page.text
```

Recordemos para hacer stemming, primero tenemos que tokenizar, asi que vamos a repetir los pasos de la tokenización.

```
import nltk

nltk.download("punkt")

tokens = nltk.word_tokenize(texto,"spanish")
```

Si lo hacen de forma local, no necesitan descargar "punkt" otra vez, pero en la nube si lo necesitamos, si ya sabemos bien qué paquete necesitamos, podemos simplemente pasarlo como parámetro.

Ahora podemos pasar por completo al stemming. NLTK también inclye herramientas para esto, dentro del módulo de `stem`.

Aquí vamos a usar el stemmer de Snowball, el mismo del que vimos las reglas en la presentación.

```
stemmer = nltk.stem.SnowballStemmer("spanish")
```

El `stemmer` es capaz de aplicar las reglas a una palabra que reciba como parámetro.

```
print(stemmer.stem("palabra"))
```

Para hacer el stemming sobre todas las palabras simplemente tenemos que hacer un recorrido sobre nuestros tokens.

```
stemms = [stemmer.stem(i) for i in tokens]
```

Veamos ahora la reducción que se pudo conseguir con este método de normalización.

```
print(len(set(tokens)))
print(len(set(stemms)))
```
