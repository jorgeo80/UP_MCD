## (1) Revisión, limpieza y clasificación de las etiquetas

Importamos la libreria **nltk**

```
import nltk
nltk.download('punkt')
nltk.download('words')
nltk.download('stopwords')
```


Generamos una función que nos ayudara a tener una base limpia y clasificada **"nlp_classification"**.

El flujo de la función será el siguiente:
1.	Limpiamos las descripciones
  - **regex0**	Quitamos todos los enters \\n los tabulares \\t y los retornos de carro \\r que pueda traer la descripción `(\n|\r|\t)`
  - **regex1**	Quitamos todos los puntos, comas y guiones dobles que estén seguidos por una letra o solos  `(\.{2,}|\,{2,}|\-{2,})([a-z])?`
  - **regex2**	Separamos todas las secuencias de dos dígitos seguidas de un punto o no \(en caso de existir el punto se remplazara por un espacio\) y una letra `(\d{2,})(\.)?([a-z])`
  - **regex3**	Separamos las secuencias de una letra o espacio con uno o más puntos o una coma seguidas o no de una letra, digito o espacio  `([a-z]|\s)(\.+|\,)([a-z]|\d|\s)?`
  - **regex4** Separamos las secuencias de una letra con un digito o un digito con una letra seguidas por un punto y una letra con un digito o un digito con una letra `([a-z]\d|\d[a-z])(\.)([a-z]\d|\d[a-z])`
  - **regex5** Separamos las secuencias un digito seguidas por un punto y una letra `(\d)(\.)([a-z])`
  - **regex6** Separamos las secuencias de un digito seguidas por un espacio un ‘x’ o una diagonal y un digito `(\d)(\s)(x|\/)(\d)`
  - **regex7** Quitamos todas las secuencias de una letra entre dos espacios `(\s)([a-z])(\s)`
  - **regex8** Separamos las secuencias de un digito con un punto y un digito seguidas de dos o más letras `(\d\.\d)([a-z]{2,})`

2.	Generamos un vector llamado **stopword_sents** que contendrá enunciados limpios sin stopwords caracteres de puntuación, no palabras o caracteres específicos
3. 	En este punto limpiamos palabras que se encuentren dentro de nuestro diccionario de equivalencias. Este diccionario contiene equivalencias de  palabras que se escriben abreviadas o con ciertos patrones específicos determinados por la institución
4.	Generamos un vector llamado **stemm_sents** que contendrá enunciados solo con stemms de cada palabra
5.	Una vez que se tienen las descripciones limpias, se corre una clasificación buscando palabras o secuencias de palabras \(regex\) claves dentro de cada una de ellas `(reloj\w{0,5}|watch)`
6.	Despues de que se corren los puntos anteriores se obtienen las siguientes salidas
	* **\(1 - Summary\)** Resumen del conteo de cada una de las clasificaciones que contiene la base a la que se le aplico el mismo
	* **\(2 - df\)**  Una base que contiene las clasificaciones, los enunciados originales, los enunciados limpios, los enunciados limpios sin stopwords y los enunciados solo con stemms.



```
class nlp_classification: 
  
  def __init__(self, df, languaje = 'spanish', dictionary = False): 
    
    # Importamos librerias necesarias

    import re
    import pandas as pd
    from nltk.tokenize import word_tokenize
    from nltk.corpus import stopwords
    from string import punctuation
    
    # Generamos un vector con los enunciados

    data = [df[i] for i in range(len(df)) if type(df[i]) == type(str())]
    
    # Generamos las regex necesarias para limpiar nuestros datos
    
    # Quitamos todos los enters \n los tabulares \t y los retornos de carro \r 
    # que pueda traer la descripción
    regex0 = re.compile(r"(\n|\r|\t)")
    # Quitamos todos los puntos, comas y guiones dobles que estén seguidos 
    # por una letra o solos
    regex1 = re.compile(r"(\.{2,}|\,{2,}|\-{2,})([a-z])?")
    # Separamos todas las secuencias de dos dígitos seguidas de un punto o no 
    # (en caso de existir el punto se remplazara por un espacio) y una letra
    regex2 = re.compile(r"(\d{2,})(\.)?([a-z])")
    # Separamos las secuencias de una letra o espacio con uno o más puntos o 
    # una coma seguidas o no de una letra, digito o espacio
    regex3 = re.compile(r"([a-z]|\s)(\.+|\,)([a-z]|\d|\s)?")
    # Separamos las secuencias de una letra con un digito o un digito con una 
    # letra seguidas por un punto y una letra con un digito o un digito con una letra
    regex4 = re.compile(r"([a-z]\d|\d[a-z])(\.)([a-z]\d|\d[a-z])")
    # Separamos las secuencias un digito seguidas por un punto y una letra 
    regex5 = re.compile(r"(\d)(\.)([a-z])")
    # Separamos las secuencias de un digito seguidas por un espacio un ‘x’ o 
    # una diagonal y un digito
    regex6 = re.compile(r"(\d)(\s)(x|\/)(\d)")
    # Quitamos todas las secuencias de una letra entre dos espacios
    regex7 = re.compile(r"(\s)([a-z])(\s)")
    # Separamos las secuencias de un digito con un punto y un digito seguidas 
    # de dos o más letras
    regex8 = re.compile(r"(\d\.\d)([a-z]{2,})")

    # Aplicamos las regex anteriores a la limpieza de los enunciados

    sents = [re.sub(regex0, r"", df[i].lower()) for i in range(len(df)) if type(df[i]) == type(str())]
    sents = [re.sub(regex1, r" \2", sent) for sent in sents]
    sents = [re.sub(regex2, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex3, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex3, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex4, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex5, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex6, r"\1\3\4", sent) for sent in sents]
    sents = [re.sub(regex7, r"\1", sent) for sent in sents]
    sents = [re.sub(regex8, r"\1 \2", sent) for sent in sents]
    sents = [sent.strip() for sent in sents]

    # Generamos una lista de stopwords

    s = " "
    customStopWords = set(stopwords.words(languaje) + list(punctuation) + ['--'])
    stopwords_sents = []

    # Corremos la lista de los enunciados limpios, corregimos palabras que se 
    # encuentren en diccionario y quitamos stopwords

    try:  
      for sent in sents:
        temp = []
        for word in nltk.word_tokenize(sent, languaje):
          if not word in customStopWords:
            if [d for d in diccionario[diccionario[diccionario.columns[0]] == word][diccionario.columns[1]]] != []:
              temp.append([d for d in diccionario[diccionario[diccionario.columns[0]] == word][diccionario.columns[1]]][0])
            else:
              temp.append(word)  
        stopwords_sents.append(s.join(temp))
    except:
      stopwords_sents = [s.join([word for word in nltk.word_tokenize(sent, languaje) if not word in customStopWords]) for sent in sents]

    # Generamos los stemms de los enunciados de nuestra base

    stemmer = nltk.stem.SnowballStemmer(languaje)
    stemms_sents = [s.join([stemmer.stem(word) for word in nltk.word_tokenize(sent, languaje) if not word in customStopWords]) for sent in sents]

    # En este punto procederemos a generar un diccionario donde incluiremos
    # los pasos anteriores mas la clasificacion por grupos

    df_groups = {'group': [],
                 'sents': [],
                 'clean_sents': [],
                 'stopwords_sents': [],
                 'stemms_sents': []}
    
    n_oro = 0
    n_reloj = 0
    n_metales_p = 0
    n_otros = 0

    for i in range(len(stopwords_sents)):
      if re.search(r'(reloj\w{0,5}|watch)', stopwords_sents[i]):
        n_reloj += 1
        df_groups['group'].append('relojes')
        df_groups['sents'].append(data[i])
        df_groups['clean_sents'].append(sents[i])
        df_groups['stopwords_sents'].append(stopwords_sents[i])
        df_groups['stemms_sents'].append(stemms_sents[i])
      elif re.search(r'(\s)?(oro|oor|roo)(\s)?', stopwords_sents[i]):
        n_oro += 1
        df_groups['group'].append('oro')
        df_groups['sents'].append(data[i])
        df_groups['clean_sents'].append(sents[i])
        df_groups['stopwords_sents'].append(stopwords_sents[i])
        df_groups['stemms_sents'].append(stemms_sents[i])
      elif re.search(r'(\s)?(plata|diamant{1,4}|gram[oia]{1}[a-z]{0,4}|grs)(\s)?', stopwords_sents[i]):
        n_metales_p += 1
        df_groups['group'].append('metales_p')
        df_groups['sents'].append(data[i])
        df_groups['clean_sents'].append(sents[i])
        df_groups['stopwords_sents'].append(stopwords_sents[i])
        df_groups['stemms_sents'].append(stemms_sents[i])  
      else:
        n_otros += 1
        df_groups['group'].append('otros')
        df_groups['sents'].append(data[i])
        df_groups['clean_sents'].append(sents[i])
        df_groups['stopwords_sents'].append(stopwords_sents[i])
        df_groups['stemms_sents'].append(stemms_sents[i])

    # Guardamos los conteos de la clasificación anterior en una variable 
    # llamada summary la cual podremos consultar si lo deseamos

    self.summary = '--- Clasifiación --- \n-------------------- \n Registros: {} \n-------------------- \n Oro: {}\n Relojes: {}\n Metales P: {}\n Otros: {}\n--------------------'.format(n_oro + n_reloj + n_metales_p + n_otros, n_oro, n_reloj, n_metales_p, n_otros)

    # Convertimos el diccionario anterior, lo convertimos en dataframe y
    # lo guardamos en una variable llamada df que podra ser usada posteriormente

    self.df = pd.DataFrame(df_groups)
```

## (3) TF-IDF y segmentación de la bolsa de palabras

Función que aplica el **tf-idf**  con el que se generarán los clusters y después  hacer el conteo de las palabras mas frecuentes y las palabras unicas por cluter

Segmentacion.JPG

El proceso de la función **nlp_cluster** se compone de los siguientes pasos:

1.	Se genera una bolsa de palabras usando la trasformación **TF-IDF**
2.	Con la bolsa de palabras obtenida del paso anterior se genera una Clusterización usando el algoritmo de **K-Means** y el número de grupos seleccionados
3.	Una vez obtenidos los clusters resultantes procedemos a hacer una Tokenización de palabras por cluster, así como el conteo de frecuencias de dicho grupo resultante
4.	En este punto seleccionaremos solo las palabras y conteos únicos por cada cluster, lo cual nos ayudara a tener una mejor vista sobre la diferenciación de palabras que se encuentran en cada cluster
5.	Guardamos los resultados en las siguientes variables:
	- 5.1 **vectorizer** Esta variable contiene el ajuste del modelo **TF-IDF** y nos ayudara posteriormente a trasformar nuevos ingresos de descripciones que se deseen modelar

 	- 5.2 **df** es la base que contiene la limpieza de las descripciones y su clasificación, a la cual añadiremos el resultado de cada cluster que se haya corrido para dicha clasificación

	- 5.3 **keywords** es el vector de palabras tokenizadas por cada cluster que se haya corrido en el modelo

	- 5.4 **keywords_counts** es la tupla de palabras con su frecuencia por cada cluster que se haya corrido en el modelo

	- 5.5 **unique_keywords** es el vector de palabras únicas que contiene cada cluster que se haya corrido en el modelo


```
class nlp_cluster: 
  
  def __init__(self, df, n_clusters = 3, init = 'k-means++', max_iter = 100, n_init = 1, verbose = True, n_top_words = 50, n_unique_words = 10, random_state=3013): 
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.cluster import KMeans
    from collections import defaultdict
    from heapq import nlargest

    vectorizer = TfidfVectorizer(max_df=0.5, min_df=2)
    X_tf_idf = vectorizer.fit_transform(df['stemms_sents'])
        
    kmeans_model = KMeans(n_clusters = n_clusters, init = init, max_iter = max_iter, n_init = n_init, verbose = verbose, random_state = random_state)
    kmeans_model.fit(X_tf_idf)

    self.vectorizer = vectorizer
    self.tf_idf = X_tf_idf
    
    df['cluster_' + str(n_clusters)] = kmeans_model.labels_
    
    self.df = df
    
    text = {}
    for i, cluster in enumerate(kmeans_model.labels_):
      oneDocument = df.loc[i, 'stopwords_sents']
      if cluster not in text.keys():
        text[cluster] = oneDocument
      else:
        text[cluster] += oneDocument
        
    keywords = {}
    keywords_counts = {}
    for cluster in range(n_clusters):
      word_sent = word_tokenize(text[cluster])
      freq = FreqDist(word_sent)
      keywords[cluster] = nlargest(n_top_words, freq, key=freq.get)
      keywords_counts[cluster] = freq
    
    self.keywords = keywords
    self.keywords_counts = keywords_counts

    unique_keys = {}
    for cluster in range(n_clusters):
      other_clusters = list(set(range(n_clusters))-set([cluster]))
      keys_other_clusters = set(keywords[other_clusters[0]]).union(set(keywords[other_clusters[1]]))
      unique = set(keywords[cluster]) - keys_other_clusters
      unique_keys[cluster] = nlargest(n_unique_words, unique, key=keywords_counts[cluster].get)

    self.unique_keywords = unique_keys
```

Creamos la función que nos ayudara a justar nuestro modelo

```
class nlp_model_fit: 
  
  def __init__(self, X, y, test_size = 0.3, random_state=3013, 
               fnModel = '',
               modelArguments = '',
               crossValidation = 10, param_grid = ''): 

    import numpy as np
    from sklearn.model_selection import train_test_split
    from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
    from sklearn.model_selection import GridSearchCV


    # Usaremos un 70%, 30% para el entrenamiento y prueba delmodelo
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state = random_state)


    # Ajustamos el modelo seleccionado
    try:
      model = fnModel(**modelArguments)
      model_cv = GridSearchCV(model, param_grid, cv = crossValidation)
      model_cv_fit = model_cv.fit(X_train,y_train)
      y_pred = model_cv_fit.predict(X_test)

      # Metricas de Validación
      acc = accuracy_score(y_test, y_pred)
      pre = precision_score(y_test, y_pred, average='micro')
      rec = recall_score(y_test, y_pred, average='micro')
      f1 = 2 * ((pre * rec) / (pre + rec))
      
      # Generamos un objeto donde guardaremos las metricas
      metrics = {
                'params': model_cv_fit.best_params_,  
                'Accuracy_train':model_cv_fit.score(X_train,y_train),
                'Accuracy_test': acc,
                'Recall_test': rec,
                'Precision_test': pre,
                'F1_test': f1
                }
      
      self.model_cv_fit = model_cv_fit
      self.metrics = metrics
      self.model = str(fnModel)
      self.model_args = modelArguments
    except:
      print('Faltan parametros')
      
```

##(5) Autoetiquetado de datos

Una vez ajustados los mejores modelos para cada uno de los grupos de clasificación y su respectivo resultado de la Clusterización, generaremos una función llamada **nlp_predict** la cual nos ayudara a generar el autoetiquetado de las nuevas descripciones. Dicha función se compone del siguiente flujo:

1.	Limpieza de las descripciones
2.	Corrección de palabras mediante el uso del diccionario de datos y eliminación de **stopwords** y signos de puntuación
3.	Creación de oraciones representadas solo por **stemms**
4.	Clasificación de descripciones por palabras o secuencias claves
5.	Transformación y creación de la bolsa de palabras usando la función de **TF-IDF**
6.	Aplicación del modelo seleccionado por cada una de las clasificaciones correspondientes, que dará como resultado el cluster esperado en dicha clasificación
7.	Obtención de los resultados mediante un **dataframe** que contiene la información de todos los puntos anteriores por cada una de las descripciones que se deseen ingresar para autoetiquetar


Autoetiquetado.JPG

```
class nlp_predict: 
  
  def __init__(self, df, languaje = 'spanish', dictionary = False, ncluster = [3,5,4,3]): 
    
    # Importamos librerias necesarias

    import re
    import pandas as pd
    from nltk.tokenize import word_tokenize
    from nltk.corpus import stopwords
    from string import punctuation
    
    # Generamos un vector con los enunciados

    data = [df[i] for i in range(len(df))]
    
    # Generamos las regex necesarias para limpiar nuestros datos

    regex0 = re.compile(r"(\n|\r|\t)")
    regex1 = re.compile(r"(\.{2,}|\,{2,}|\-{2,})([a-z])?")
    regex2 = re.compile(r"(\d{2,})(\.)?([a-z])")
    regex3 = re.compile(r"([a-z]|\s)(\.+|\,)([a-z]|\d|\s)?")
    regex4 = re.compile(r"([a-z]\d|\d[a-z])(\.)([a-z]\d|\d[a-z])")
    regex5 = re.compile(r"(\d)(\.)([a-z])")
    regex6 = re.compile(r"(\d)(\s)(x|\/)(\d)")
    regex7 = re.compile(r"(\s)([a-z])(\s)")
    regex8 = re.compile(r"(\d\.\d)([a-z]{2,})")

    # Aplicamos las regex anteriores a la limpieza de los enunciados

    sents = [re.sub(regex0, r"", data[i].lower()) for i in range(len(data))]
    sents = [re.sub(regex1, r" \2", sent) for sent in sents]
    sents = [re.sub(regex2, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex3, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex3, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex4, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex5, r"\1 \3", sent) for sent in sents]
    sents = [re.sub(regex6, r"\1\3\4", sent) for sent in sents]
    sents = [re.sub(regex7, r"\1", sent) for sent in sents]
    sents = [re.sub(regex8, r"\1 \2", sent) for sent in sents]
    sents = [sent.strip() for sent in sents]

    # Generamos una lista de stopwords

    s = " "
    customStopWords = set(stopwords.words(languaje) + list(punctuation) + ['--'])
    stopwords_sents = []

    # Corremos la lista de los enunciados limpios, corregimos palabras que se 
    # encuentren en diccionario y quitamos stopwords

    try:  
      for sent in sents:
        temp = []
        for word in nltk.word_tokenize(sent, languaje):
          if not word in customStopWords:
            if [d for d in diccionario[diccionario[diccionario.columns[0]] == word][diccionario.columns[1]]] != []:
              temp.append([d for d in diccionario[diccionario[diccionario.columns[0]] == word][diccionario.columns[1]]][0])
            else:
              temp.append(word)  
        stopwords_sents.append(s.join(temp))
    except:
      stopwords_sents = [s.join([word for word in nltk.word_tokenize(sent, languaje) if not word in customStopWords]) for sent in sents]

    # Generamos los stemms de los enunciados de nuestra base

    stemmer = nltk.stem.SnowballStemmer(languaje)
    stemms_sents = [s.join([stemmer.stem(word) for word in nltk.word_tokenize(sent, languaje) if not word in customStopWords]) for sent in sents]

    # En este punto procederemos a generar un diccionario donde incluiremos
    # los pasos anteriores mas la clasificacion por grupos

    df_test = {
               'sents': [],
               'clean_sents': [],
               'stopwords_sents': [],
               'stemms_sents': [],
               'group_val': [],
               'cluster_val': []
               }    
    
    for i in range(len(stopwords_sents)):
      if re.search(r'(reloj\w{0,5}|watch)', stopwords_sents[i]):
        df_test['sents'].append(data[i])
        df_test['clean_sents'].append(sents[i])
        df_test['stopwords_sents'].append(stopwords_sents[i])
        df_test['stemms_sents'].append(stemms_sents[i])
        df_test['group_val'].append('relojes')
        # == Predicción Cluster Relojes
        stemm = [stemms_sents[i]]
        X_temp = eval('cluster_relojes_' + str(ncluster[0]) + '.vectorizer.transform(stemm)')
        X_temp = model_relojes.model_cv_fit.predict(X_temp)[0]
        df_test['cluster_val'].append(X_temp)
      elif re.search(r'(\s)?(oro|oor|roo)(\s)?', stopwords_sents[i]):
        df_test['sents'].append(data[i])
        df_test['clean_sents'].append(sents[i])
        df_test['stopwords_sents'].append(stopwords_sents[i])
        df_test['stemms_sents'].append(stemms_sents[i])
        df_test['group_val'].append('oro')
        # == Predicción Cluster Oro
        stemm = [stemms_sents[i]]
        X_temp = eval('cluster_oro_' + str(ncluster[1]) + '.vectorizer.transform(stemm)')
        X_temp = model_oro.model_cv_fit.predict(X_temp)[0]
        df_test['cluster_val'].append(X_temp)
      elif re.search(r'(\s)?(plata|diamant{1,4}|gram[oia]{1}[a-z]{0,4}|grs)(\s)?', stopwords_sents[i]):
        df_test['sents'].append(data[i])
        df_test['clean_sents'].append(sents[i])
        df_test['stopwords_sents'].append(stopwords_sents[i])
        df_test['stemms_sents'].append(stemms_sents[i])
        df_test['group_val'].append('metales_p')
        # == Predicción Cluster Metales P
        stemm = [stemms_sents[i]]
        X_temp = eval('cluster_metales_p_' + str(ncluster[2]) + '.vectorizer.transform(stemm)')
        X_temp = model_metales_p.model_cv_fit.predict(X_temp)[0]
        df_test['cluster_val'].append(X_temp)
      else:
        df_test['sents'].append(data[i])
        df_test['clean_sents'].append(sents[i])
        df_test['stopwords_sents'].append(stopwords_sents[i])
        df_test['stemms_sents'].append(stemms_sents[i])
        df_test['group_val'].append('otros')
        # == Predicción Cluster Otros
        stemm = [stemms_sents[i]]
        X_temp = eval('cluster_otros_' + str(ncluster[3]) + '.vectorizer.transform(stemm)')
        X_temp = model_otros.model_cv_fit.predict(X_temp)[0]
        df_test['cluster_val'].append(X_temp)

    self.df = pd.DataFrame(df_test)
```
