```
import spacy

pln = spacy.load("es_core_news_sm")
```

```
import requests

url = 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/M%C3%A9xico_texto.txt'
page = requests.get(url)
texto = page.text
```

```
doc = pln(texto)
```

```
for oracion in doc.sents:
  print(oracion.text)
  print("="*20)
```

```
for token in doc:
  print(token)
```

```
for token in doc:
  print(token,token.is_stop,sep="\t\t")
```

```
for token in doc:
  print(token,token.pos_,sep="\t\t")
```

```
for token in doc:
  print(token,token.lemma_,sep="\t\t")
```

```
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

oracion = nlp("Apple is looking at buying U.K. startup for $1 billion")

displacy.render(oracion,style="dep",jupyter=True)
```

```
for token in oracion:
  print(token.text, token.dep_, token.head.text, [child for child in token.children],sep="\t\t")
```
