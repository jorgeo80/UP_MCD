Limpiar el texto:

```
import requests

url = 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/ejercicioRE.txt'
page = requests.get(url)
texto = page.text
```

```
print(texto)
```

Importamos la libreria re

```
import re
```

Generamos el REGEX que nos limpie los tags del texto

```
regex = re.compile(r"&?(html|lt|[\]]);?[\w\/\!\[\]\-]{0,9}([\s]href=[\"].+[\"])?(&gt;)?")
print(re.sub(regex,'',texto))
```
