```
import requests
 
url = 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/IslaMisteriosa.txt'
page = requests.get(url)
texto_train = page.text
 
url = 'https://raw.githubusercontent.com/apimentelaUP/recursos/master/PuebloAereo.txt'
page = requests.get(url)
texto_test = page.text
 
len(texto_test)
```

La tarea es evaluar nuestro modelo de lenguaje. Recordemos nuestro modelo de lenguaje:

```
!pip install sentence_splitter
 
import collections
import nltk
from sentence_splitter import SentenceSplitter, split_text_into_sentences
 
nltk.download("punkt")
 
splitter = SentenceSplitter(language='es')
oraciones_train = splitter.split(texto_train)
 
oraciones_tokenizadas_train = []
for oracion in oraciones_train:
  tokens = nltk.word_tokenize(oracion,"spanish")
  oraciones_tokenizadas_train.append(tokens)
 
for i,oracion in enumerate(oraciones_tokenizadas_train):
  oraciones_tokenizadas_train[i] = ["<s>"] + oracion + ["</s>"]
 
unigramas_oraciones_train = []
bigramas_oraciones_train = []
 
for oracion in oraciones_tokenizadas_train:
  unigramas = nltk.ngrams(oracion, 1) 
  bigramas = nltk.ngrams(oracion, 2)
 
  unigramas_oraciones_train.append(unigramas)
  bigramas_oraciones_train.append(bigramas)
 
unigramas_train = [j for i in unigramas_oraciones_train for j in i]
bigramas_train = [j for i in bigramas_oraciones_train for j in i]
 
frecuencia_unigramas_train = collections.Counter(unigramas_train)
frecuencia_bigramas_train = collections.Counter(bigramas_train)
```

```
oraciones_test = splitter.split(texto_test)

oraciones_tokenizadas_test = []
for oracion in oraciones_test:
  tokens =nltk.word_tokenize(oracion,"spanish")
  oraciones_tokenizadas_test.append(tokens)
```

```
for i,oracion in enumerate(oraciones_tokenizadas_test):
  oraciones_tokenizadas_test[i] = ["<s>"] + oracion + ["</s>"]
```

```
bigramas_oraciones_test = []

for oracion in oraciones_tokenizadas_test:
  bigramas = nltk.ngrams(oracion,2)
  bigramas_oraciones_test.append(bigramas)
```

```
bigramas_test = [j for i in bigramas_oraciones_test for j in i]
bigramas_test
```

```
import math

suma_logs = 0
N = 0

for bigrama in bigramas_test:
  numerador = frecuencia_bigramas_train[bigrama]
  denominador = frecuencia_unigramas_train[bigrama[:-1]]

  if not numerador or not denominador : continue

  # La manera correcta:
  # - Sumarle uno a todo
  # - Etiquetas de palabras raras [UNK]

  probabilidad = numerador/denominador

  suma_logs += math.log(probabilidad)
  N += 1

print(suma_logs)
print(N)
```

```
perplejidad = math.exp(-suma_logs/N)
print(perplejidad)
```
