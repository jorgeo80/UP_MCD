Vamos a comenzar con un texto, Python los distingue gracias a las comillas, pueden ser comillas dobles o sencillas, e incluso pueden ser dobles triples o sencillas triples, si es que el texto puede contener comillas también.

```
texto='''

    Mi hermano Alberto cayó al pozo cuando tenía cinco años.

    Fue una de esas tragedias familiares que sólo alivian el tiempo y la circunstancia de la familia numerosa.

    Veinte años después mi hermano Eloy sacaba agua un día de aquel pozo al que nadie jamás había vuelto a asomarse.

    En el caldero descubrió una pequeña botella con un papel en el interior.

    "ESTE ES UN MUNDO COMO OTRO CUALQUIERA", decía el mensaje.

'''
```

En python, las variables que contienen texto se pueden analizar como si fueran listas de caracteres. Si queremos saber qué tan largo es un texto, podemos hacer lo siguiente:

```
len(texto)
```

Python también cuenta con una serie de herramientas básicas para manipular texto.
Operaciónes comunes como:

-  Fragmentar
- Concacetar
- Buscar
- Reemplazar

Entre otras funciones propias de los textos. Por ejemplo, Python nos puede indicar si una cadena está formada por minúsculas, mayúsculas o números. Pero para poder trabajar cómodamente, es mejor primero fragmentar el texto. Python ofrece la función split(), que recibe como parámetro el caracter delimitador para hacer la segmentación.

```
segmentado = texto.split(" ")
print(segmentado)
len(segmentado)
```

```
# Palabras con mas de 5 letras
largas = [x for x in segmentado if len(x) > 5]
# Palabras mayúsculas
mayusculas = [x for x in segmentado if x.isupper()]
# Palabras que comienzan con mayúscula
primera_mayuscula = [x for x in segmentado if x.istitle()]
```

```
print(largas)
print(mayusculas)
print(primera_mayuscula)
```

Ya que el texto se puede ver como una lista de caracteres, podemos hacer búsquedas de texto con `in`. Tal cual lo haríamos para una lista normal.
Además de eso, Python tiene funciónes para buscar al inicio o al final de un texto.



```
# Palabras que contienen "el"
con_el = [x for x in segmentado if "el" in x]
# Palabras que comienzan con "a"
primera_a = [x for x in segmentado if x.startswith("a")]
# Palabras que terminan con "s"
ultima_s = [x for x in segmentado if x.endswith("s")]
```

```
print(con_el)
print(primera_a)
print(ultima_s)
```

Otras funciones similares incluyen:
- `s.islower()`
- `s.isalpha()`
- `s.isdigit()`
- `s.isalnum()`

Pasando a la tarea de modificar texto, podemos encontrar funciones para convertir las letras a mayúsculas o minúsculas, para unir listas de palabras en textos mas largos y por supuesto para reemplazar texto. 

```
# Texto en minusculas
minusculas = texto.lower()
# Texto reunificado
reunificado = " | ".join(segmentado)
# Reemplazando 4 espacios por tabulaciones
tabulaciones = texto.replace("    ","\t")
```

```
print(minusculas)
print(reunificado)
print(tabulaciones)
```

Muchas veces, solo queremos buscar un texto dentro de otro, no necesariamente para reemplazarlo.

```
loc_hermano = texto.find("hermano")
print(texto[loc_hermano:loc_hermano+len("hermano")])

locr_hermano = texto.rfind("hermano")
print(texto[:locr_hermano])
```

Seguramente ya habían notado esos espacios al inicio, para muchas aplicaciones ese tipo de formatos no es de interés y pueden ser molestos, por ejemplo, a la hora de segmentar. Es por eso que Python también tiene funciones para limpiar el texto.

```
# Nueva segmentación por salto de línea
segmentado_nl = texto.splitlines()
# Limpieza de texto
lineas_limpias = [ x.strip() for x in segmentado_nl if x ]
```

```
print(segmentado_nl)
print(lineas_limpias)
```
