Cuando se trabaja con Python de forma local, la manera mas sencilla de trabajar con archivos es tenerlos en un directorio de trabajo conocido, por lo regular suele ser el mismo en el que se tiene el código.

```
# La 'w' es para indicar que el archivo se abre en modo escritura
#  si ese archivo tuviera contenido previo, se sobreescribe
with open("archivo_texto.txt",'w') as f:
  f.write("Aquí podemos escribir texto hacia el archivo")
  # No hay que olvidar el salto de línea
  f.write("\n")
  f.write("Y podemos poner mas texto\n")
  lines = ["Tambien podemos tener","una lista","con varias lineas\n","pero también necesitan su salto de línea\n"]
  f.writelines(lines)
```

El archivo que se acaba de crear, se creó en la máquina virtual de Colab, lo podemos abrir nuevamente y ver su contenido, pero hay que tener muy claro que estos archivos no están conectados con los locales ni con los de drive.

```
# La 'r' es para indicar que el archivo se abre en modo lectura
with open("archivo_texto.txt",'r') as f:
  contenido = f.read()
  print(contenido)
```

Es importante tener en cuenta que el archivo está en la maquina virtual, de hecho podemos obtener una lista de los archivos que existen en nuestra carpeta de trabajo.

```
from os import listdir

# Aquí podremos observar nuestro nuevo archivo
listdir()
```

Para que se muestren de forma local los archivos que tenemos en el Drive, es necesario montarlo, la forma mas sencilla de lograrlo es a través del ícono de carpeta que está del lado izquiero de la página; allí se tiene un botón con la opción "Activar unidad de Drive".

Esto logrará que aparezca una carpeta nueva con todos los datos que se tengan en el drive.

```
from os import listdir

# Aquí podremos observar la nueva carpeta llamada "drive"
listdir()
```

Otra opción para poder "leer" archivos fuera de la máquina virtual de Colab, y que también funciona de forma local, es mediante Github. Por ejemplo, podemos usar la liga: <https://raw.githubusercontent.com/apimentelaUP/corpora/master/trabalenguas.txt>

```
import requests
url = 'https://raw.githubusercontent.com/apimentelaUP/corpora/master/trabalenguas.txt'
page = requests.get(url)
texto = page.text
print(texto)
```

Si se activa la unidad de Drive, uno puede guardar archivos directamente en la nube, pero otra opción es descargar los archivos de la máquina virtual de Colab directamente a la máquina local. Una manera de lograrlo es mediante el navegador de carpeta de la barra lateral izquierda, hay que actualizarla. Otra forma es directamente con código.

```
from google.colab import files

# Con esta instrucción se le indica cuál archivo es el que se debe descargar
files.download('archivo_texto.txt') 
```

Para subir archivos locales a Colab se tienen las mismas dos opciones

```
from google.colab import files

# Con esta instrucción se pide un archivo al usuario
info_subida = files.upload() 
```
