Importamos la libreria NLTK

```
import nltk
nltk.download("punkt")
```

Separamos los versos y tokenizamos cada uno de ellos

```
versos_tokenizados = []
for verso in versos.split('\n'):
  tokens = nltk.word_tokenize(verso,"spanish")
  versos_tokenizados.append(tokens)
```

Añadimos espacios al principio y al final de cada uno de los versos

```
for i, verso in enumerate(versos_tokenizados):
  versos_tokenizados[i] = ["<s>"] + verso + ["</s>"]
#print(versos_tokenizados)
```

Creamos un vector con los unigramas y los bigramas

```
unigramas_versos = []
bigramas_versos = []

for verso in versos_tokenizados:
  unigramas = nltk.ngrams(verso, 1)
  bigramas = nltk.ngrams(verso, 2)

  unigramas_versos.append(unigramas)
  bigramas_versos.append(bigramas)

for gramas in unigramas_versos[0]:
  print(gramas)
```

```
unigramas = [j for i in unigramas_versos for j in i]
#print(unigramas)
bigramas = [j for i in bigramas_versos for j in i]
#print(bigramas)
```

Procesamos las frecuencias

```
import collections

frecuencia_unigramas=collections.Counter(unigramas)
frecuencia_bigramas=collections.Counter(bigramas)
```

Ejecutamos el modelo

```
from nltk.probability import ConditionalFreqDist
from nltk.tokenize.treebank import TreebankWordDetokenizer

versos_completos = []
for verso in versos_incompletos.split('\n'):
  #print("Original:--" + verso)
  tokens = nltk.word_tokenize(verso,"spanish")
  if(len(tokens) > 0):
    condicion  = (tokens[-1],)
    frecuencias_condicionales = ConditionalFreqDist()

    for unigrama in frecuencia_unigramas.keys():
      frecuencia = frecuencia_bigramas[condicion + unigrama]
      if frecuencia:
        frecuencias_condicionales[condicion][unigrama] = frecuencia
      cont = frecuencias_condicionales[condicion].N()
    if cont > 1:
      palabra_mayor = frecuencias_condicionales[condicion].max() 
    else:
      palabra_mayor = frecuencias_condicionales[condicion]
    tokens.append(str(palabra_mayor[0]))
    oracion = TreebankWordDetokenizer().detokenize(tokens)
      #print("Completada:" + oracion)
    versos_completos.append(oracion)
```

```
#Limpiamos el valor <s> ya que este resultado es la ingesta para el siguiente
versos_completos_final = []
for verso in versos_completos:
  verso = verso.replace('<s>', '')
  verso = verso.replace('</s>', '')
  versos_completos_final.append(verso)
```
