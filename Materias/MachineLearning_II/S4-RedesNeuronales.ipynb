```
from keras.models import Sequential
from keras.layers import Dense
from keras import backend as k
```

```
k.backend() # Con esto podemos ver que framework esta utilizando
```

### Karma de Machine Learning
1. Modelo
2. Perdida
3. Optimizacion

### Uso de la funcion model.add() refiere a la capa que se va a correr
1. Dense quiere decir que todas las unidades se contectan con la capa siguiente
2. El primer parametro es el numero de neuronas en la capa
3. input_dim: el numero de variables que se quiere uar
4. activation: la funcion de activacion que se va a usar

##### * [Resnet](http://personal.cimat.mx:8181/~mrivera/cursos/aprendizaje_profundo/resnet/resnet.html) ejemplo de arquitectura de red neuronal con funciones de activacion diferentes


```
model = Sequential() # Usa las capas una detras de otra

#Función de Activación
activationFunction = 'sigmoid'

# Primera capa
model.add(Dense(12, input_dim = 8, activation = activationFunction)) 
# Segunda capa
model.add(Dense(10, activation = activationFunction))
# Capa de Salida
model.add(Dense(1, activation = activationFunction))
```


Tensorflow no hace nada hasta que el modelo es compilado

Uso de la función **model.compile**:
1. Loos: Función de perdida que se usara
2. Optimizer: algoritmo de optimización de parámetros
3. Metrics: metrica que usara para optimizar la perdida

```
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd', metrics = ['accuracy'] )
```

Importamos la base de datos **Pima Indians Diabetes Database**

```
import pandas as pd
df = pd.read_csv('diabetes.csv')
df.head()
```

Separamos el vector de respuesta Y y la matriz de variables X

```
X = df[df.columns[:-1]]
Y = df['Outcome']
```

Ajustamos el modelo

```
model.fit(X,Y, epochs = 300, batch_size = 10)
```

Imprimimos los Scores del modelo seleccionado

```
scores = model.evaluate(X, Y)
scores
```
