```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Visualizaci√≥n de datos:

```python
x = np.array([5,15,30,25,35,45,47,55]).reshape((-1,1))
y = np.array([5,20,18,14,32,22,32,38])

plt.scatter(x,y)
```

```python
def train_linear(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    xy_dot = np.dot(x.T,y)
    n      = len(x)
    
    w0 = (y_mean*sum(x**2) - x_mean*xy_dot)/(sum(x**2) - n*x_mean**2) 
    w1 = (xy_dot - n*x_mean*y_mean)/(sum(x**2) - n*x_mean**2)
    
    return w0,w1
```

```python
print(train_linear(x,y))
```

```python
def predict_linear(x,w):
    return w[0] + w[1]*x 
```

```python
w = train_linear(x,y)
y_pred = predict_linear(x,w)

x_test = 20
y_test = predict_linear(x_test,w)

plt.scatter(x,y,color='black')
plt.plot(x,y_pred,marker='*')
plt.plot(x_test,y_test,marker='o')
```

```python
model = LinearRegression()
model.fit(x,y)
y_sk = model.predict(x)

print(model.intercept_)
print(model.coef_)

plt.plot(x,y_sk,marker='*')
plt.scatter(x,y)
```
